```python
import numpy as np
from time import time

from keras.datasets import mnist

from tmu.models.classification.vanilla_classifier import TMClassifier

(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

X_train = np.where(X_train.reshape((X_train.shape[0], 28*28)) > 75, 1, 0)
X_test = np.where(X_test.reshape((X_test.shape[0], 28*28)) > 75, 1, 0)

clauses=500
s=7.5
T=10 
epochs=20

tm = TMClassifier(clauses,T, s, max_included_literals=16, platform='CPU', weighted_clauses=True,incremental='True')


print("\nAccuracy over %d epochs:\n",epochs)
for i in range(epochs):
        start_training = time()
        tm.fit(X_train, Y_train)
        stop_training = time()

        start_testing = time()
        result = 100*(tm.predict(X_test) == Y_test).mean()
        stop_testing = time()

        print("#%d Accuracy: %.2f%% Training: %.2fs Testing: %.2fs" % (i+1, result, stop_training-start_training, stop_testing-start_testing))
        
        total_class=0
        
        for i in range (10):
                total_clause=0
                for j in range (clauses):
                        k=tm.number_of_include_actions(i,j)
                        total_clause=total_clause+k
                total_class=total_class+total_clause
                     
        
        includes=((total_class*100)/(1568*clauses*10))          
        print("Total Includes: %f \n" % (includes) )



```
